```python
import datafaucet as dfc
import datafaucet.logging as log 
```

```python
dfc.project.load('test')
```

```python
engine = dfc.project.engine()
spark = engine.context()
```

### Fullscan copy

```python
md = dfc.project.metadata()
path = md['resources'].keys()[0]

md_src = dfc.project.resource(path, 'source')
md_trg = dfc.project.resource(path, 'target')

md_src.update({'date_window':None, 'date_start':None, 'date_end':None})
md_trg.update({'date_window':None, 'date_start':None, 'date_end':None})

engine.copy(md_src,md_trg,'overwrite')
```

### Update copy

```python
md_src = dfc.resource(path, 'source')
md_trg = dfc.resource(path, 'target')

engine.copy(md_src,md_trg)
```

```python
engine.stop()
```
